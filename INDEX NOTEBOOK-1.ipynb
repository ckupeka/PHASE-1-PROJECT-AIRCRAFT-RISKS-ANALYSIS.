## 4. Data Cleaning.

### 4.1 Missing values and dropping columns.
First we will check for the missing values.
Then we will sort the missing values.
We will perform a arithmetic operation of multiplying the missing values by 100, in order to convert them to percentage form, then we will proceed to drop the columns with missing values of 20% and above.

```python
# Checking for missing values.
missing_df=df.isnull().sum()
print(type(missing_df))
print(missing_df)
```

```python
#Sorting the missing values and converting them to percentage form.
sorted_df = missing_df.sort_values(ascending=False)
total_rows = df.shape[0]
missing_df_percentage = (missing_df / total_rows) * 100
missing_df_percentage = missing_df_percentage.sort_values(ascending=False).round(0)

# Identifying the columns with more than 20% missing values
dropped_columns = missing_df_percentage > 20.0

# Printing the columns with missing values greater than 20%
print("\nMissing Values Percentage > 20:\n", missing_df_percentage[dropped_columns])

# Printing the missing values count for each column
print("\nMissing Values Count:\n", sorted_df)

# Dropping the columns with missing values above 20%
df_cleaned = df.drop(columns=missing_df_percentage[missing_df_percentage > 20].index)

# Displaying the cleaned dataframe
print("\nDataframe after dropping columns with >20% missing values:\n", df_cleaned.head())

```

Next we are going to check the missing values in the remaining columns, then divide the columns to categorical columns(non-numerical data) and numerical columns(numerical data) then we are going to fill the null values in each column differently. The Categorical column will be filled with the value 'Unknown' and the numerical columns with the value '0'.

```python
#In order to fill the null or missing values appropriately we check the all the column data type
#First we check for all the numerical columns 
numerical_columns = df_cleaned.select_dtypes(include=['number']).columns
print(numerical_columns)
```

As we can see the numerical columns include ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries',
       'Total.Minor.Injuries', 'Total.Uninjured']. So we will proceed to fill them with the value '0'. I chose '0' because i did not want there to be any change o my analysis.

```python
 # Replacing Null values in numerical columns with 0
numerical_colums= ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries',
       'Total.Minor.Injuries', 'Total.Uninjured']
for col in numerical_columns:
    df_cleaned[col] = df_cleaned[col].fillna(0)  # Replacing Null values with 0
    print(df_cleaned[col])
```

```python
#Then we check for the categorical columns.
categorical_columns = df_cleaned.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

As we can see these are the categorical columns include ['Investigation.Type', 'Event.Date', 'Location', 'Country',
       'Aircraft.damage', 'Make', 'Model', 'Engine.Type', 'Weather.Condition'],
     

```python
#Filling the Null values in the categorical columns with the value 'Unknown'
categorical_columns = ['Investigation.Type', 'Event.Date', 'Location', 'Country',
       'Aircraft.damage', 'Make', 'Model', 'Engine.Type', 'Weather.Condition']
for col in categorical_columns:
    df_cleaned[col] = df_cleaned[col].fillna('Unknown')  
    print(df_cleaned[col])
```

Next we are going to all the data in the categorical columns to uppercase.

```python
for column in ['Investigation.Type', 'Event.Date', 'Location', 'Country',
       'Aircraft.damage', 'Make', 'Model', 'Engine.Type', 'Weather.Condition']:
    df_cleaned[column] = df_cleaned[column].str.upper()
```

We will proceed with the merging of the USState Codes with the Aviation data set, and we will merge them through the location column.

```python
#First we will also change the abbreviations in the USState Codes to uppercase to  for easy merging with the Aviation data set.
for column in ["US_State"]:
    usc[column] = usc[column].str.upper()
```

```python
usc.head()
```

```python
# Extract the state code from the 'Location' column
# Split the 'Location' string by commas
# Select the second part (index 1) of the split string which should be the state code
# Remove any leading or trailing whitespace from the extracted string
df_cleaned['State Code']=df_cleaned['Location'].str.split(',').str[1].str.strip()

# Remove any leading or trailing whitespace from the 'Abbreviation' column in the usc  DataFrame
usc['Abbreviation']=usc['Abbreviation'].str.strip()
```
